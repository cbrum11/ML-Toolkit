# Helpful Pandas Data Cleanup

The following is a general pipeline for cleaning/exploring data in Pandas

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
df = pd.read_csv('datasets/iris.data')
```

```python
df.head()
```

## Add column headers

```python
columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']
df.columns = columns
df.head()
```

## Get Data Snapshot

```python
df.describe()
```

## Get Entries Per Label/Category

```python
df_copy = df.groupby('class').size()
print(df_copy)
```

# Visualization

## Pair Plot

```python
sns.pairplot(df, hue='class', height=5);
```

## Histogram

```python
df.hist(edgecolor='black', linewidth=1.2, figsize=(12,8))
plt.show()
```

## Violin Plot

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.violinplot(x='class',y='sepal length', data=df)
plt.subplot(2,2,2)
sns.violinplot(x='class',y='sepal width', data=df)
plt.subplot(2,2,3)
sns.violinplot(x='class',y='petal length', data=df)
plt.subplot(2,2,4)
sns.violinplot(x='class',y='petal length', data=df)
```

## Box Plot

```python
df.boxplot(by='class', figsize=(12,8))
```

```python
pd.plotting.scatter_matrix(df, figsize=(12,8))
plt.show()
```
